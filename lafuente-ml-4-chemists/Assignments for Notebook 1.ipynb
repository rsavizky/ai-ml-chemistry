## Excercise 1.7: Scatter plot/Regression of random numbers - error bars - Matplotlib
This will be your most comprehensive task so far, you'll need to use many libraries.
* Generate two lists of random numbers, one for the X values of your plot and one for the Y values and sort them in ascending order.
* Plot your list of values, put labels and a title.
* Calculate a linear regression and show the equation on the plot.
* Calculate error bars and put them on the plot.

##### Hints:
* Use the native python function `random` to generate random numbers. Other useful Python functions are `range` and `random.sample` .

More information on `random` function [here](https://docs.python.org/3/library/random.html). 

More information for `range` function [here](https://www.w3schools.com/python/ref_func_range.asp)
* You can find information on sorting values [here](https://www.w3schools.com/python/ref_list_sort.asp)

Solution:
##### First, let's automatically generate some values: 

````python
import random
#Generate 20 random numbers between 10 and 100
randomlist_x = random.sample(range(1, 60), 20)
print(randomlist_x)
````

These will be our "x" values.
Now let's generate another list of 20 random numbers between 20 and 100 to be our "y" values and name it *randomlist_y*:

````python
#Generate 20 random numbers between 1 and 100
randomlist_y = random.sample(range(1, 60), 20)
print(randomlist_y)
````

Let's use the **sort()** function on our list of x values to sort the values in ascending order. This is a built- in  Python function. If you don't enter any additional parameters, it sorts the elements in the list in ascending order. This function changes our list directly, so there's no need to define a new sorted list. Also visualize the new version of the lists by printing them, in order to make sure the values are sorted correctly.

````python
randomlist_x.sort()
randomlist_y.sort()

print(randomlist_x)
print(randomlist_y)
````

Let's import **numpy** and **matplotlib.pyplot** modules for our plot, and let's define our first list of numbers as our "x" values. After that do the same procedure on the second list to redefine the set of "y" values:


````python
import matplotlib.pyplot as plt
import numpy as np

x = randomlist_x
y = randomlist_y
````

Let's make a scatter plot with these values. Use the *scatter()* function of matplotlib. 
You can add a title to your graph witht he *title()* function, and label the axes with the xlabel() and ylabel() functions respectively.

````python
plt.scatter(x,y, marker='o')
plt.title("Scatter Plot")
plt.xlabel('x values')
plt.ylabel("y values")
````

Now let's draw a trendline over the previous graph by defining a new variable as the 1-degree polynomial fit of your distribution. On numpy this function is called "polyfit", and the parameters should be x, y, and the degree of the polinomial you want to fit your distribution to.

In order to make a more undestandable graph, it's better to add "legends", to label everything properly. Add an additional "label" parameter to the scatter() function, and then use the  legend() function to show these legends on the graph.

````python
plt.scatter(x, y, label="scatter")

z = np.polyfit(x, y, 1) # z gives the slope and intercept as it's output on a list
p = np.poly1d(z) #p will give the "y" values correponding to the x values.
plt.plot(x,p(x),"r-", label="trendline")

plt.title("Scatter Plot and trendline")
plt.xlabel("x values")
plt.ylabel("y values")
plt.legend()
plt.show()
````

If we want to show the function equation, we can use: 


```
print("The curve fit for this function is y = {0} x + {1}.".format(z[0], z[1]))
```
"z" being the fit of the 1D polynomial. z[0] the slope of the adjusted function and z[1] the intercept.


````python
print("The curve fit for this function is y = {0:1.4E} x + {1:1.4E}.".format(z[0], z[1]))
````

Now let's draw some error bars. Use the *plt.errorbar()* function. You will need to add another variable: the y-error, which will be our linear regression minus the given "y" values. The parameters for your plt.errorbar() function will be these three variables.

````python
plt.scatter(x, y, label="scatter")
plt.xlabel("x values")
plt.ylabel("y values")
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r-", label="trendline")
yerr=abs(p(x)-y) # Armaan: ABS() added to fix "ValueError: 'yerr' must not contain negative values"
plt.errorbar(x, p(x), yerr, fmt='.k',label="error")

plt.rc('font', size=18)          # controls default text sizes
plt.rc('axes', titlesize=18)     # fontsize of the axes title
plt.rc('axes', labelsize=18)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=18)    # fontsize of the tick labels
plt.rc('ytick', labelsize=18)    # fontsize of the tick labels
plt.rc('legend', fontsize=14)    # legend fontsize
plt.rc('figure', titlesize=22)  # fontsize of the figure title

plt.legend()
plt.show() 
print("The curve fit for this function is y = {0:1.4E} x + {1:1.4E}.".format(z[0], z[1]))
````
